```python
import zipfile
from pathlib import Path
import pandas as pd

zip_path = Path(r"C:\Users\Ayush\Downloads\DATA3888\data3888 finc6\Optiver.zip")
extract_dir = Path(r"C:\Users\Ayush\Downloads\DATA3888\data3888 finc6\Optiver_extracted")

if not extract_dir.exists():
    with zipfile.ZipFile(zip_path, "r") as zf:
        zf.extractall(extract_dir)
    print("Extracted successfully")
else:
    print("Already extracted")

csv_files = list(extract_dir.rglob("*.csv"))
print(f"Total CSV files found: {len(csv_files)}")

converted_count = 0
failed_count = 0

for file in csv_files:
    try:
        df = pd.read_csv(file)
        out_file = file.with_suffix(".parquet")
        df.to_parquet(out_file, engine="pyarrow", index=False)
        converted_count += 1
        print(f"Converted: {file.name}")
    except Exception as e:
        failed_count += 1
        print(f"Failed on {file.name}: {e}")

parquet_files = list(extract_dir.rglob("*.parquet"))

print("\nSummary")
print(f"CSV files found: {len(csv_files)}")
print(f"Successfully converted: {converted_count}")
print(f"Failed: {failed_count}")
print(f"Parquet files now present: {len(parquet_files)}")
```
